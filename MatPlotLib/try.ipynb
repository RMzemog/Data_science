```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel("Temperature")
plt.title('Weather')
plt.plot(x,y,linestyle = 'dotted')
```

```python
plt.plot(x,y,'D')
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label ='Max')
plt.plot(days,min_t, label = 'Mix')
plt.plot(days,avg_t, label = 'Avg')
plt.legend(loc = 'best', shadow=True,fontsize = 'large')
plt.grid()
```

```python
company =['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos = np.arange(len(company))
plt.ylabel('revenue(bin)')
plt.title("US Tech Stocks")
plt.xticks(ypos,company)
plt.bar(ypos,revenue, label= 'Revenue')
plt.bar(ypos,profit, label = 'Profit')
plt.legend()
```

```python

plt.title("US Tech Stocks")
plt.xticks(ypos,company)
plt.barh(ypos,revenue, label= 'Revenue')
plt.barh(ypos,profit, label = 'Profit')
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins= [80,100,125,150], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'], orientation='horizontal')
plt.legend()
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilites']
```

```python
plt.pie(exp_vals, labels = exp_labels, autopct= '%0.1f%%', shadow=True, explode =[0,0.1,0.1,0,0], startangle=180)
plt.savefig("piechart.png")
plt.show()
```
