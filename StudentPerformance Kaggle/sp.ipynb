```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("StudentsPerformance.csv")
print(df.head())
```

```python
male = df.loc[df["gender"] == 'male'].gender.count()
female = df.loc[df['gender'] == 'female'].gender.count()
genderr = [male,female]
lab = ['male','female']
print(genderr)
```

```python
plt.pie(genderr, autopct= '%0.1f%%', startangle= 90)
plt.show()
```

```python
#plt.pie(exp_vals, labels = exp_labels, autopct= '%0.1f%%', shadow=True, explode =[0,0.1,0.1,0,0], startangle=180
dd = df['race/ethnicity'].value_counts()
dd = dd.reset_index()
dd.columns = ['groups', 'race']
print(dd)
hh = list(dd.groups)
hhh = list(dd.race)
print(hh)
print(hhh)
```

```python
plt.pie(hhh, labels=hh, autopct='%0.1f%%',explode=[0,0,0,0,0.2],shadow= True)
plt.show()
```

```python
ee = df['parental level of education'].value_counts()
ee = ee.reset_index()
ee.columns = ['colleges','totals']
print(ee)
```

```python
plt.xticks(rotation='vertical',size=9)
plt.barh(ee.colleges,width=ee.totals)
plt.show() 
```
