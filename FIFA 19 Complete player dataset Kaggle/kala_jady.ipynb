```python
def valuem(mm):
    mm = mm.replace('€',"")
    if 'M' in mm:
        mm = mm.replace("M","")
        mm = float(mm) * 1000000
    elif 'K' in mm:
        mm = mm.replace("K","")
        mm = float(mm) * 1000
    print(mm)
    return mm
```

```python
df['Value'] = df['Value'].apply(lambda x: valuem(x))
```

```python
print(df.Value.head())
```

```python
def mandola(mn):
    mn = str(mn)
    if(mn == ""):
        return 0
    if len(mn) > 4:
        mn = mn[-2:]
    else:
        mn = int(mn)/100
    return int(mn)
```

```python
df['Contract Valid Until'] = df['Contract Valid Until'].apply(lambda x: mandola(x))
```
