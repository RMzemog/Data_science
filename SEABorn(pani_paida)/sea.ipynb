```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.barplot(x = 'year', y = 'passengers', data = flights)
```

```python
flights.head()
```

```python
dd = flights.groupby(['year'], sort = True).sum().idxmax()
ddd = flights.groupby(['year']).sum()
print(ddd)
year = []
month = []
for i, j in ddd.items():
    for k,l in j.items():
        year.append(k)
        month.append(l)
plt.plot(year, month)
```

```python
dddd = flights.groupby('year').passengers.sum()
sns.barplot(dddd.index, dddd.values)
```

```python
sns.barplot(x = 'month', y = 'passengers',palette = 'winter_r', data = flights)
```
