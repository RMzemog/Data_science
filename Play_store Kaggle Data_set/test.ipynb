```python
#df.loc[df["paid"] == 'False'].title
x = df.loc[df['paid'] == True].count(axis = 0).title
y = df.loc[df['paid'] == False].count(axis = 0).title
z = [x,y]
plt.pie(z, labels = ['True','False'],autopct='%0.1f%%')
plt.show()
```
