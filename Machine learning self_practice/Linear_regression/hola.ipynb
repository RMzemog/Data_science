```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('train.csv')
df.head()
df2 = pd.read_csv('test.csv')
```

```python
df.isnull().sum()
df.dropna(how='any',inplace=True)
```

```python
X = df.iloc[: ,:1].values
y = df.iloc[: ,1:].values
print(X)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X,y)
```
